###3.6 Checking and treating missing values

```python
# Check for missing values in each column
missing_values = data.isnull().sum()
print(missing_values)
```


*     **Gender** is a categorical variable for which there is no numerical average (like mean or median) to calculate.
*     **Partner_Salary** is a numerical variable, representing a continuous value (e.g., salary).



```python
# Filling missing values in 'Gender' with the mode (most frequent value)
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])

# Filling missing values in 'Partner_salary' based on mean within each 'Profession' group
data['Partner_salary'] = data.groupby('Profession')['Partner_salary'].transform(lambda x: x.fillna(x.mean()))
```



*   As **Gender** is a categorical variable (for which there is no numerical average (like mean or median) to calculate), we can fill the missing values with the mode, by assuming that the most common category is a reasonable estimate for missing entries.
*   As **Partner_Salary** is a numerical variable filling missing values with the mean salary within each profession helps maintain a more accurate representation of the data and minimize distortion of salary values compared to simply using the overall mean, which might not capture the differences in salary distribution across professions.


```python
print("Missing values in each column:\n", data.isnull().sum())
```



*    All the missing values have been successfully dealt with as both outputs show zero missing values for relevant columns.

###4.1.4 **Dealing With Outliers**

```python
# Calculate Q1 and Q3 and define bounds for capping
Q1 = data['Total_salary'].quantile(0.25)
Q3 = data['Total_salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap Total Salary at the calculated bounds
data['Total_salary_capped'] = np.where(
    data['Total_salary'] > upper_bound, upper_bound,
    np.where(data['Total_salary'] < lower_bound, lower_bound, data['Total_salary'])
)

```

```python
# Plot the boxplot of Total_salary after capping
plt.figure(figsize=(6, 4))
sns.boxplot(data['Total_salary_capped'], color='blue')
plt.title('Boxplot of Total Salary (After Capping Outliers)')
plt.xlabel('Total Salary (Capped)')
plt.show()
```
